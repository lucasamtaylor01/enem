```python
import pandas as pd
from pathlib import Path
import pandas as pd
from pathlib import Path
from sklearn.preprocessing import StandardScaler
```

```python
data_dir = Path("data")
INDIR = Path("data/data_raw")
OUTDIR = Path("data/data_processed")

OUTDIR.mkdir(parents=True, exist_ok=True)
```

```python
arquivo = INDIR / "PARTICIPANTES_2024.csv"
df = pd.read_csv(arquivo, sep=';', encoding='latin-1')
```

```python
df_simplificado = df[['TP_FAIXA_ETARIA', 'TP_SEXO', 'TP_ESTADO_CIVIL', 'TP_COR_RACA', 'IN_TREINEIRO', 'SG_UF_PROVA', 'Q007']]
```

```python
faixa_etaria_map = {
    1: 16,
    2: 17,
    3: 18,
    4: 19,
    5: 20,
    6: 21,
    7: 22,
    8: 23,
    9: 24,
    10: 25,
    11: (26+30)/2,
    12: (31+35)/2,
    13: (36+40)/2,
    14: (41+45)/2,
    15: (46+50)/2,
    16: (51+55)/2,
    17: (56+60)/2,
    18: (60+80)/2,
    19: (60+80)/2,
    20: (60+80)/2
}


df_simplificado = df_simplificado.rename(columns={'TP_FAIXA_ETARIA': 'IDADE'})
df_simplificado['IDADE'] = df_simplificado['IDADE'].map(faixa_etaria_map)
df_simplificado['IDADE'] = df_simplificado['IDADE'].astype('int64')
```

```python
cor_raca_map = {
    0: "NÃO DECLARADO",
    1: "BRANCA",
    2: "PRETA",
    3: "PARDA",
    4: "AMARELA",
    5: "INDÃGENA",
    # 6: "NÃO DISPÃE DA INFORMAÃÃO" -> AGRUPADO COM 0 POR SIMPLIFICAÃÃO
    6: "NÃO DECLARADO"
}

df_simplificado['TP_COR_RACA'] = df_simplificado['TP_COR_RACA'].map(cor_raca_map)
df_simplificado = df_simplificado[df_simplificado['TP_COR_RACA'] != 'NÃO DECLARADO']
```

```python
df_simplificado = df_simplificado[df_simplificado['IN_TREINEIRO'] != 1]
df_simplificado = df_simplificado.drop(columns=['IN_TREINEIRO'])
```

```python
cor_renda_map = {
    "A": 0,
    "B": 1412/2,
    "C": (1412+2118)/2,
    "D": (2118+2824)/2,
    "E": (2824+3530)/2,
    "F": (3530+4236)/2,
    "G": (4236+5648)/2,
    "H": (5648+7060)/2,
    "I": (7060+8472)/2,
    "J": (8472+9884)/2,
    "K": (9884+11296)/2,
    "L": (11296+12708)/2,
    "M": (12708+14120)/2,
    "N": (14120+16944)/2,
    "O": (16944+21180)/2,
    "P": (21180+28240)/2,
    "Q": (28240+35000)/2
}

df_simplificado['Q007'] = df_simplificado['Q007'].map(cor_renda_map)
df_simplificado['Q007'] = df_simplificado['Q007'].astype("Float64")
df_simplificado = df_simplificado.rename(columns={'Q007': 'RENDA_FAMILIAR'})
```

```python
estado_civil_map = {
    0: "NAO_INFORMADO",
    1: "SOLTEIRO",
    2: "UNIAO",
    3: "DIVORCIADO",
    4: "NAO_INFORMADO" # Simplificando
}

df_simplificado['TP_ESTADO_CIVIL'] = df_simplificado['TP_ESTADO_CIVIL'].map(estado_civil_map)
df_simplificado = df_simplificado[df_simplificado['TP_ESTADO_CIVIL'] != 'NAO_INFORMADO']
```

```python
caminho_tratado = OUTDIR / "PARTICIPANTES_SIMPLIFICADO_TRATADO.csv"
df_simplificado.to_csv(caminho_tratado, index=False)
```

```python
df_modelagem = pd.get_dummies(
    df_simplificado,
    columns=['TP_ESTADO_CIVIL', 'TP_COR_RACA', 'TP_SEXO']
)
```

```python
scaler = StandardScaler()
df_modelagem[['RENDA_FAMILIAR', 'IDADE']] = scaler.fit_transform(df_modelagem[['RENDA_FAMILIAR', 'IDADE']])
```

```python
caminho_modelo = OUTDIR / "PARTICIPANTES_SIMPLIFICADO_TRATADO_MODELO.csv"
df_modelagem.to_csv(caminho_modelo, index=False)
```
